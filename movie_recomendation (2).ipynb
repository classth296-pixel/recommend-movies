```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
#
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)

movies
```

MAKING SIMILAR WORDS IN TO A SINGLE WORD (EXA--LOVED->LOVE,ACTIVITIES->ACTIVITY)

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
stem(new_df['tags'][1])
```

```python
new_df['tags']=new_df['tags'].apply(stem)
new_df
```
